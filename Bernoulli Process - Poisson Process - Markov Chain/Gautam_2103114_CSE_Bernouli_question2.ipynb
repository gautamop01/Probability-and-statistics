```python
# Roll. N. 2103114
# Importing libraries
import random
from matplotlib import pyplot as plt

prob,t=0.5,10000  # Here prob is probability and t is trials  
str_s = [] #string size

for i in range(t): # Trials to get k arrivals
  s = 0 # size initialise value 0
  while (random.uniform(0,1) > prob):
    pass
  while (random.uniform(0,1) < prob):
    s += 1
    s += 1
  str_s += [s]
print(str_s)
set_str_s = set(str_s)
dict_str = {} # constructing the string dictionary and adding values to the plot graph
for i in set_str_s:
  dict_str[i] = 0
for i in str_s:
  dict_str[i] += 1/t

print(dict_str)

# Now, graphing the data
x_axis= list(dict_str.keys())  # list all arrival values on X axis
y_axis= list(dict_str.values()) # list all arrival values on Y axis
plt.bar(x_axis,y_axis,color='orange',edgecolor='black',width=0.8)
plt.title("Empirical Distribution") # Give Title Name to Graph
plt.xlabel("Length of string") # Print "Length of string " on x axis
plt.ylabel("Probability") # Print "probability" on y axis
plt.show()

#---> Theoritical Geometric Process <----
pass_t_f = [] # passes trial to first 
for i in range(t): # creating a list of the successful test runs for the graph
  k = 0
  while (random.uniform(0,1) > prob):
    k += 1
  k += 1
  pass_t_f += [k]
set_pass_t = set(pass_t_f)
pass_t_prob = {}
for i in set_pass_t: # putting the values of successful test results into a dictionary to plot the graph
  pass_t_prob[i] = 0
for i in pass_t_f:
  pass_t_prob[i] += 1/t

# Now, graphing the data
x_axis = list(pass_t_prob.keys()) # list all arrival values on X axis
y_axis = list(pass_t_prob.values()) # list all arrival values on y axis
plt.bar(x_axis,y_axis,color='orange', edgecolor='black', width=0.8)
plt.title("Geometric Distribution") # Give Title Name to Graph
plt.xlabel("Passed Trial")
plt.ylabel("Probability") # Print "probability" on y axis
plt.show()

```
